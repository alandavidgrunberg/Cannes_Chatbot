```python
import pandas as pd
```

```python
# need to update/make new csvs because wikipedia added more movies/info
```

```python
palm = pd.read_csv("../csvs/updated/raw/palm.csv")
certain = pd.read_csv("../csvs/updated/raw/certain.csv")
out = pd.read_csv("../csvs/updated/raw/out.csv")
series = pd.read_csv("../csvs/updated/raw/series.csv")
midnight = pd.read_csv("../csvs/updated/raw/midnight.csv")
premiere = pd.read_csv("../csvs/updated/raw/premiere.csv")
special = pd.read_csv("../csvs/updated/raw/special.csv")
shorts = pd.read_csv("../csvs/updated/raw/shorts.csv")
critics = pd.read_csv("../csvs/updated/raw/critics.csv")
critics_special = pd.read_csv("../csvs/updated/raw/critics_special.csv")
directors = pd.read_csv("../csvs/updated/raw/directors.csv")
directors_special = pd.read_csv("../csvs/updated/raw/directors_special.csv")
```

```python
raw_dfs = [palm, certain, out, series, midnight, premiere, special, shorts, \
           critics, critics_special, directors, directors_special]

for df in raw_dfs:
    df.rename(columns={'Production country':'Production Countrie(s)'},inplace=True)
    df['Categorie(s)'] = ''

raw_dfs[0]
```

```python
### creating full_df from raw dfs

full_df = pd.concat(raw_dfs, ignore_index=True).drop_duplicates()
full_df
```

```python
### creating new df of Caméra d'Or eligibe films

camera = palm[palm['English title'].str.contains('(CdO)')]
camera2 = certain[certain['English title'].str.contains('(CdO)')]
camera3 = out[out['English title'].str.contains('(CdO)')]
camera4 = series[series['English title'].str.contains('(CdO)')]
camera5 = midnight[midnight['English title'].str.contains('(CdO)')]
camera6 = premiere[premiere['English title'].str.contains('(CdO)')]
camera7 = special[special['English title'].str.contains('(CdO)')]
camera8 = shorts[shorts['English title'].str.contains('(CdO)')]
camera9 = critics[critics['English title'].str.contains('(CdO)')]
camera10 = critics_special[critics_special['English title'].str.contains('(CdO)')]
camera11 = directors[directors['English title'].str.contains('(CdO)')]
camera12 = directors_special[directors_special['English title'].str.contains('(CdO)')]

camera = pd.concat([camera,camera2,camera3,camera4,camera5,camera6,camera7,\
                    camera8,camera9,camera10,camera11,camera12])

camera
```

```python
complete_df = full_df.copy()

palm_mask = full_df.isin(palm.to_dict('list')).all(axis=1)
complete_df.loc[palm_mask, "Categorie(s)"] += "Competing for Palme d'Or, "

certain_mask = full_df.isin(certain.to_dict('list')).all(axis=1)
complete_df.loc[certain_mask, "Categorie(s)"] += "Competing in Un Certain Regard, "

out_mask = full_df.isin(out.to_dict('list')).all(axis=1)
complete_df.loc[out_mask, "Categorie(s)"] += "Out of competition, "

series_mask = full_df.isin(series.to_dict('list')).all(axis=1)
complete_df.loc[series_mask, "Categorie(s)"] += "Television series, "

midnight_mask = full_df.isin(midnight.to_dict('list')).all(axis=1)
complete_df.loc[midnight_mask, "Categorie(s)"] += "Midnight screening, "

premiere_mask = full_df.isin(premiere.to_dict('list')).all(axis=1)
complete_df.loc[premiere_mask, "Categorie(s)"] += "Cannes Premiere, "

special_mask = full_df.isin(special.to_dict('list')).all(axis=1)
complete_df.loc[special_mask, "Categorie(s)"] += "Special Screening, "

shorts_mask = full_df.isin(shorts.to_dict('list')).all(axis=1)
complete_df.loc[shorts_mask, "Categorie(s)"] += "Short film, "

critics_mask = full_df.isin(critics.to_dict('list')).all(axis=1)
#complete_df.loc[critics_mask, "Categorie(s)"] += "International Critics' Week: in competition, "
complete_df.loc[critics_mask, "Categorie(s)"] += "International Critics' Week, "

critics_special_mask = full_df.isin(critics_special.to_dict('list')).all(axis=1)
#complete_df.loc[critics_special_mask, "Categorie(s)"] += "International Critics' Week: special screening, "
complete_df.loc[critics_special_mask, "Categorie(s)"] += "International Critics' Week, "

directors_mask = full_df.isin(directors.to_dict('list')).all(axis=1)
#complete_df.loc[directors_mask, "Categorie(s)"] += "Directors' Fortnight: in competition, "
complete_df.loc[directors_mask, "Categorie(s)"] += "Directors' Fortnight, "

directors_special_mask = full_df.isin(directors_special.to_dict('list')).all(axis=1)
#complete_df.loc[directors_special_mask, "Categorie(s)"] += "Directors' Fortnight: special screening, "
complete_df.loc[directors_special_mask, "Categorie(s)"] += "Directors' Fortnight, "

camera_mask = full_df.isin(camera.to_dict('list')).all(axis=1)
complete_df.loc[camera_mask, "Categorie(s)"] += "Eligible for Caméra d'Or, "

```

```python
complete_df
```

```python
# only run once, to get rid of trailing ',' in Categorie(s) column vals

complete_df['Categorie(s)'] = complete_df['Categorie(s)'].str.slice(stop=-2)

complete_df
```

```python
for index, row in complete_df.iterrows():
    if '(CdO)' in row['English title']:
        complete_df.at[index, 'English title'] = row['English title'][:-6]
    if '(CdO)' in row['Original title']:
        complete_df.at[index, 'Original title'] = row['Original title'][:-6]
    if '(opening film)' in row['English title']:
        complete_df.at[index, 'English title'] = row['English title'][:-15]
    if '(opening film)' in row['Original title']:
        complete_df.at[index, 'Original title'] = row['Original title'][:-15]
    if '(closing film)' in row['English title']:
        complete_df.at[index, 'English title'] = row['English title'][:-15]
    if '(closing film)' in row['Original title']:
        complete_df.at[index, 'Original title'] = row['Original title'][:-15]
        
complete_df
```

```python
complete_df = complete_df.reset_index(drop=True)
complete_df
```

```python
complete_df.to_csv("../chatbot_trials/trial2/csvs/complete_data.csv",index=False)
```
